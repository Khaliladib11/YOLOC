```python
from bdd import BDD
from pathlib import Path
import matplotlib.pyplot as plt
```

```python
objects_to_detect = ['pedestrian', 'rider', 'truck', 'bus', 'motorcycle', 'bicycle', 'car', 'traffic light']

bdd_params = {
    "image_root": Path('../../dataset'),
    "detection_classes": objects_to_detect,
    "stage": 'train',
    "det_task": True,
    "de_seg_task": True,
    "ll_seg_task": True,
}

bdd = BDD(**bdd_params)
```

```python
len(bdd.images)
```

```python
annotations = bdd.get_detection_annotation(0)
```

```python
image = bdd.get_image(0)
```

```python
import matplotlib.patches as patches
```

```python
plt.imshow(image)
```

```python
annotations
```

```python
annotations
```

```python
len(annotations['boxes'])
```

```python
fig, ax = plt.subplots()
ax.imshow(image)

for i in range(len(annotations['boxes'])):
    b = annotations['boxes'][i]
    rect = patches.Rectangle((b[0], b[1]), b[2], b[3], facecolor="none", edgecolor='r')
    ax.add_patch(rect)

plt.axis('off')
plt.show()
```
