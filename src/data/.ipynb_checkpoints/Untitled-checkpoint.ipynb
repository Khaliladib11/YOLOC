# Class

```python
from torch.utils import data
from collections import deque
from PIL import Image
```

```python
class BDD(data.Dataset):
    
    def __init__(self, root, objects_to_detect, stage='train', tasks=['detection', 'drivable_area_segmentation', 'lane_segmentation']):
        
        
        assert stage in ['train', 'test'], 'stage has to be in ["train", "test"]'
        for task in tasks:
            assert task in ['detection', 'drivable_area_segmentation', 'lane_segmentation'], 'task has to be in ["detection", "drivable_area_segmentation", "lane_segmentation"]'
        
        self.root = root
        self.stage = stage
        self.tasks = tasks
        self.objects_to_detect = objects_to_detect
        
        self.image_path = os.path.join(self.root, 'images/bdd100k/images/100k', self.stage)
        
        self.det_annotations = os.path.join(self.root, 'det_annotations', self.stage)
        self.da_seg_annotations = os.path.join(self.root, 'da_seg_annotations', self.stage)
        self.ll_seg_annotations = os.path.join(self.root, 'll_seg_annotations', self.stage)
        
        self.images = os.listdir(self.image_path)
        
        
        
    def get_annotation(self, idx):
        
        assert idx < len(self.images) or idx < 0, 'index out of range'
        image = self.images[idx]
        
        annotations = {}
        
        if 'detection' in self.tasks:
            
            detection_annotations = deque()
            
            file = open(os.path.join(self.det_annotations, image.split('.')[0]+'.json'))
            
            annotation = json.load(file)
            objects = annotation['frames'][0]['objects']
            for obj in objects:
                if obj['category'] in self.objects_to_detect:
                    if obj['category'] == 'traffic light':
                        cls = "traffic light" + " " + obj['attributes']['trafficLightColor']
                        bbox = obj['box2d']
                        detection_annotations.append([cls, bbox])
                        
                    else:
                        cls = obj['category']
                        bbox = obj['box2d']
                        detection_annotations.append([cls, bbox])
                
            annotations['detection'] = detection_annotations
            
            file.close()
            
            
        return annotations
    
    
    def load_image(self, idx):
        image = Image.open(os.path.join(self.image_path, self.images[idx]))
        return image
    
    
    def display_image(self, idx):
        image = self.load_image(idx)
        detection_annotation = self.get_annotation(idx)['detection']
        for obj in detection_annotation:
            print(obj)
        plt.imshow(image)
        plt.axis('off')
        
        
        # plot the image
        fig, ax = plt.subplots()
        ax.imshow(image)
        for i in range(len(detection_annotation)):
            bbox = detection_annotation[i][1]
            
            rect = patches.Rectangle((bbox['x1'], bbox['y1']), bbox['x2'] - bbox['x1'], bbox['y2'] - bbox['y1'],
                                     edgecolor=color_map[class_],
                                     facecolor="none")
            ax.add_patch(rect)
            break

        plt.axis('off')
        plt.show()

            
    
    def __len__(self):
        return len(self.images)
    
    def __getitem__():
        pass
```

```python
objects_to_detect = ['pedestrian', 'rider', 'truck', 'bus', 'motorcycle', 'bicycle', 'car', 'traffic light']
bdd_params = {
    "root": '../../dataset',
    "objects_to_detect": objects_to_detect,
    "stage": 'train',
    "tasks": ['detection', 'drivable_area_segmentation', 'lane_segmentation']
}

bdd = BDD(**bdd_params)
```
